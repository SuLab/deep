## what is the `calibration_results` database table/view?

     bucket | num_variables | num_correct | num_incorrect 
    --------+---------------+-------------+---------------
          0 |         21087 |          17 |           285
          1 |        107692 |         304 |          2174

          9 |         10603 |         618 |            53
    (10 rows)


```python
rng = np.arange(0, 1, 0.1)

print('\t'.join(('bucket', 'num_variables', 'num_correct', 'num_incorrect')))
for i, lower_bound in enumerate(rng):
    df_123_subset = df_123[(df_123['expectation'] > lower_bound) & (df_123['expectation'] <= lower_bound+0.1)]
    print(i, end='\t')
    print(len(df_123_subset), end='\t\t')
    is_correct_false, is_correct_true = df_123_subset['is_correct'].value_counts()
    print(is_correct_true, end='\t\t')
    print(is_correct_false)
```

**`bucket`** column represents a decile for expectation values in a certain range
*(0,0.1], (0.1, 0.2], ...*

**`calibration_results`** view has columns **`num_correct`**, **`num_incorrect`** that correspond to the number of values with `row['is_correct'] == 't'` or `row['is_correct'] == 'f'`
